<h2>Define surface and photon properties </h2>

We assume the crystal is tightly wrapped or painted, so the index of refraction is irrelevant. Set the inside and outside the same. The reflection of the surface has two components. One is random reflection. The reflected photon is uniformly distributed over the entire semi-hemisphere. The second component is diffused reflection. The reflected photon follows the law of reflection smeared by a Gaussian distribution. 

If the photon hits a sensor, it follows the Fresnel's equations of reflection and transmission http://en.wikipedia.org/wiki/Fresnel_equations. The reflection part is calculated using diffused reflection mentioned above.

Here we assume the following,

For photons reflecting on crystal surface/wrapping/painting material:<br>
Random reflaction probability = 0.1<br>
Diffused reflection probability = 0.89<br>
$\sigma$ of diffused reflection = 22 degrees<br>
Ignore the index of refraction

For photons reflecting in the sensor area<br>
Diffused reflection probability = 1.0<br>
$\sigma$ of diffused reflection = 1 degrees<br>
Assume photons going from crystal (LYSO $n=1.82$ or BaF2 $n=1.47$) to glass ($n= 1.52$).


```python
# Surface property
crys_surf= dict(random_reflect=0.1, diffuse_reflect=0.89, diffuse_sigma=22, idx_refract_in=1.5, idx_refract_out=1.5)
lyso_glass= dict(random_reflect=0.0, diffuse_reflect=1.0, diffuse_sigma=1, idx_refract_in=1.82, idx_refract_out=1.52)
bf2_glass= dict(random_reflect=0.0, diffuse_reflect=1.0, diffuse_sigma=1, idx_refract_in=1.47, idx_refract_out=1.52)
lyso_air= dict(random_reflect=0.0, diffuse_reflect=1.0, diffuse_sigma=1, idx_refract_in=1.82, idx_refract_out=1.0)
bf2_air= dict(random_reflect=0.0, diffuse_reflect=1.0, diffuse_sigma=1, idx_refract_in=1.47, idx_refract_out=1.0)
```

<h2>Define geometries</h2>
Crystals<br>
Hexagon: length= 11 cm, edge-to-edge width= 3.2 cm<br>
Square: length= 11 cm, side= 3 cm<br>

Sensors<br>
Two 1 cm$\times$0.5 cm photo sensors


```python
# Hexagon
ghex= hex_prism([0,0,0], length=11, width=3.2, **crys_surf)
# Rectangle
grect= rect_prism([0,0,0], dim=[1.5,1.5,5.5], **crys_surf)
# Sensors
location1= [0,0.5,-5.5]
location2= [0,-0.5,-5.5]
xlen, ylen=  1.0, 0.5
sen1_lg = rectangle(location1, xlen=xlen, ylen=ylen, sensitive=True, **lyso_glass)
sen2_lg = rectangle(location2, xlen=xlen, ylen=ylen, sensitive=True, **lyso_glass)
sen1_bg = rectangle(location1, xlen=xlen, ylen=ylen, sensitive=True, **bf2_glass)
sen2_bg = rectangle(location2, xlen=xlen, ylen=ylen, sensitive=True, **bf2_glass)
sen1_la = rectangle(location1, xlen=xlen, ylen=ylen, sensitive=True, **lyso_air)
sen2_la = rectangle(location2, xlen=xlen, ylen=ylen, sensitive=True, **lyso_air)
sen1_ba = rectangle(location1, xlen=xlen, ylen=ylen, sensitive=True, **bf2_air)
sen2_ba = rectangle(location2, xlen=xlen, ylen=ylen, sensitive=True, **bf2_air)
```

```python
# Define crystals
chex_lg= Crystal('chex_lg', ghex+[sen1_lg, sen2_lg])
chex_bg= Crystal('chex_bg', ghex+[sen1_bg, sen2_bg])

crec_lg= Crystal('crec_lg', grect+[sen1_lg, sen2_lg])
crec_bg= Crystal('crec_bg', grect+[sen1_bg, sen2_bg])

chex_la= Crystal('chex_la', ghex+[sen1_la, sen2_la])
chex_ba= Crystal('chex_ba', ghex+[sen1_ba, sen2_ba])

crec_la= Crystal('crec_la', grect+[sen1_la, sen2_la])
crec_ba= Crystal('crec_ba', grect+[sen1_ba, sen2_ba])
```

<h2>Run experiments</h2>

```python
def run_exp(crystal, zpoints, dz, dr, nperz, mfp=1000):
    '''
    Run experiments. Return photon detection efficiency (and error) for each z points. 
    Photons are generated in a region specified by zpoints, dz (range in z (+-dz)) and dr (a circle with radius dr in x-y plane). 
    Directions are random in 4pi.
    Return value is an array of efficiencies and an array of errors
    *crystal*: an instance of Crystal class
    *zpoints*: an array of the z coordinates
    *nperz*: number of photons per z location
    *mfp*: mean free path of the photon
    '''
    effs, errs= [], []
    for z in zpoints:
        #print z
        ndet=0
        for i in range(nperz):
            xp,rp= generate_p6(crystal.cog, dz, dr)
            photon= Photon(xp,rp, mfp= mfp)
            plane= photon.propagate(crystal)
            if photon.status != 'transmitted': continue
            if photon.lastplane is None: continue
            if photon.lastplane.sensitive:
                ndet= ndet+1
            
        ef= ndet/float(nperz)
        er= np.sqrt( ef*(1-ef) / float(nperz) )
        effs.append(ef)
        errs.append(er)
    return np.array(effs), np.array(errs)
```

```python
nprnd.seed(0)
results= {}
kwargs= dict(zpoints= np.linspace(-5,5,11), dz= 0.2, dr= 1.5, nperz= 1000)
starttime= time.time()
for crystal in [chex_la, chex_lg, chex_ba, chex_bg, crec_la, crec_lg, crec_ba, crec_bg]:
    print 'processing ', crystal.name, 
    effs, errs= run_exp(crystal, **kwargs)
    results[crystal.name+'_effs']= effs
    results[crystal.name+'_errs']= errs
    print '  time %.1f s' % (time.time() - starttime )
```

```python
import pickle
pickle.dump(results, open('eff_result.p', 'wb'))
```

```python
def eff_vs_z(ax, xp, results, key, **kwargs):
    y= results[key+'_effs']
    ey = results[key+'_errs']
    ax.errorbar(xp, y, ey, **kwargs)
```

```python
fig= figure(figsize=(16,6))
xp= np.linspace(0.5,10.5,11)
ax= plt.subplot(121)
eff_vs_z(ax, xp, results, 'chex_lg', fmt='bo-', label='LYSO-Glass')
eff_vs_z(ax, xp, results, 'chex_la', fmt='go-', label='LYSO-Air', mfc='w', mec='g', mew=2)
eff_vs_z(ax, xp, results, 'chex_bg', fmt='rx--', label='BaF2-Glass', mew=2)
eff_vs_z(ax, xp, results, 'chex_ba', fmt='cd--', label='BaF2-Air', mfc='w', mec='c', mew=2)
ax.set_title('Hexagon', fontsize='x-large')
ax.set_xlabel('Distance from sensor face', fontsize='x-large')
ax.set_ylabel('Efficiency')
ax.set_ylim(0,0.55)
ax.grid()
ax.legend(numpoints=1, framealpha=0.5)

ax= plt.subplot(122)
eff_vs_z(ax, xp, results, 'crec_lg', fmt='bo-', label='LYSO-Glass')
eff_vs_z(ax, xp, results, 'crec_la', fmt='go-', label='LYSO-Air', mfc='w', mec='g', mew=2)
eff_vs_z(ax, xp, results, 'crec_bg', fmt='rx--', label='BaF2-Glass', mew=2)
eff_vs_z(ax, xp, results, 'crec_ba', fmt='cd--', label='BaF2-Air', mfc='w', mec='c', mew=2)
ax.set_title('Square', fontsize='x-large')
ax.set_xlabel('Distance from sensor face', fontsize='x-large')
ax.set_ylabel('Efficiency')
ax.set_ylim(0,0.55)
ax.grid()
ax.legend(numpoints=1, framealpha=0.5)
```
